```python
import pandas as pd
import json
from pandas.io.json import json_normalize

with open('C:/Users/phili/Downloads/COH_ZIPCODE.geojson') as f:
    zipjson=json.load(f)

cohzips=json_normalize(zipjson['features'])
cohzips.drop(['type','properties.OBJECTID_1','properties.OBJECTID','properties.CREATION_U','properties.DATE_CREAT','properties.DATE_MODIF',
            'properties.DATE_POSTE','properties.LAST_USER','properties.SYS_ID','properties.GlobalID','properties.Shapearea','properties.Shapelen',
            'geometry.type'],axis=1,inplace=True)
cohzips.columns=['zipcode','polygons']
cohzips[['zipcode','polygons']]=cohzips[['zipcode','polygons']].astype(str)
cohzips['polygons']=cohzips['polygons'].str.replace(r'\[\[\[','[[')
cohzips['polygons']=cohzips['polygons'].str.replace(r'\]\]\]',']]')
```
