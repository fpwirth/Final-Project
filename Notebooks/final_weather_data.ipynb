```python
#Load dependencies and define data files to import
from datetime import datetime as dt, timedelta
import datetime
import pandas as pd
import numpy as np

```

```python

weather_data = pd.read_excel("../Clean Data Files/Weather_KIAHstation_2017-2020.xls")
weather_data
```

```python
weather_data['date_field_str'] =  weather_data["Year"].astype(str) + '-' + weather_data["Month"].astype(str) + '-' + weather_data["dt"].astype(str)
weather_data['date_field'] = weather_data['date_field_str'].apply(pd.to_datetime,errors='coerce')
weather_data['date_field_str'] = weather_data['date_field'].dt.strftime('%Y%m%d')
weather_data['date_field'] = weather_data['date_field'].dt.date

weather_data
```

```python
weather_data_selected = weather_data[['date_field', 'tempMax', 'tempAvg', 'tempMin', 'precTotal']]
weather_data_selected = weather_data_selected.rename(columns={"precTotal": "precipitation"})
weather_data_selected
```

```python
weather_data_selected.to_csv("../static/data/Final_selected_weather_data.csv")
```

# lets have the data ready for visualization
    

```python
weather_data = pd.read_csv("../../WeatherUnderground_KIAH_3yrs.csv")
weather_data
```

```python
weather_data['date_field_str'] =  weather_data["Year"].astype(str) + '-' + weather_data["Month"].astype(str) + '-' + weather_data["dt"].astype(str)
weather_data['date_field'] = weather_data['date_field_str'].apply(pd.to_datetime,errors='coerce')
weather_data['date_month'] = weather_data['date_field'].dt.strftime('%b')
weather_data['date_field_str'] = weather_data['date_field'].dt.strftime('%Y%m%d')
weather_data['yr'] = weather_data['date_field'].dt.strftime('%Y')
weather_data['date_field'] = weather_data['date_field'].dt.date
weather_data
```

```python
weather_data_selected = weather_data[['date_field', 'date_month', 'date_field_str', 'yr', 'tempMax', 'tempAvg', 'tempMin','dewMax', 'dewAvg', 'dewMin','humMax', 'humAvg', 'humMin','presMax', 'presAvg', 'presMin', 'precTotal']]
weather_data_selected = weather_data_selected.rename(columns={"precTotal": "precipitation"})
weather_data_selected
```

```python
weather_data_selected.to_csv("../static/data/selected_weather_data_for_visual.csv", index=False)
```

```python
weathAgg = weather_data_selected.groupby(['date_month']).agg(
    prec=('precipitation', 'sum'),
    avg_max=('tempMax', 'mean'))
```
