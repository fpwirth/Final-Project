```python
#Load dependencies and define data files to import
from datetime import datetime as dt, timedelta
import datetime
import pandas as pd
import numpy as np

url2017='https://hfdapp.houstontx.gov/311/311-Public-Data-Extract-2017.txt'
url2018='https://hfdapp.houstontx.gov/311/311-Public-Data-Extract-2018.txt'
url2019='https://hfdapp.houstontx.gov/311/311-Public-Data-Extract-2019.txt'
url2020='https://hfdapp.houstontx.gov/311/311-Public-Data-Extract-monthly.txt'
```

```python
#Create 2017 data frames
data2017=pd.read_csv(url2017,header=5,sep='|',error_bad_lines=False)


```

```python
weather_data = pd.read_excel("../Clean Data Files/Weather_KIAHstation_2017-2020.xls")
weather_data
```

```python
weather_data['date_field_str'] =  weather_data["Year"].astype(str) + '-' + weather_data["Month"].astype(str) + '-' + weather_data["dt"].astype(str)
```

```python
weather_data['date_field'] = weather_data['date_field_str'].apply(pd.to_datetime,errors='coerce')
```

```python
weather_data['date_field'] = weather_data['date_field'].dt.date
```

```python
weather_data
```

```python
data2017=data2017.drop(data2017.index[0]).reset_index(drop=True)
data2017.columns=['case number','sr location','county','district','neighborhood','tax id','trash quad','recycle quad','trash day','heavy trash day','recycle day','key map',
                'management district','department','division','sr type','queue','sla','status','sr create date','due date','date closed','overdue','title','x','y','latitude',
                'longitude','channel type']
data2017[['sr create date','due date','date closed']]=data2017[['sr create date','due date','date closed']].apply(pd.to_datetime,errors='coerce')
data2017
```

```python
data2017['date_field'] = data2017['sr create date'].dt.date 
```

```python
alldata2017 = data2017.merge(weather_data,  on="date_field")
```

```python
alldata2017.head(10)
```

```python
alldata2017["time_to_complete_hrs"] = (alldata2017["date closed"]-alldata2017["sr create date"]).astype('timedelta64[h]')
alldata2017["time_to_complete_days"] = (alldata2017["time_to_complete_hrs"] /24)
alldata2017["time_to_complete_days"] = alldata2017["time_to_complete_days"].apply(np.ceil)
```

```python
alldata2017.to_csv("../clean Data Files/Hou311andWeather2017.csv", index=False)
```
